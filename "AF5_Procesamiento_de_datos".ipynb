```python
#1._Cargar el dataset con pandas
import pandas as pd

robcsv = pd.read_csv('robot_sensors.csv')

```

```python
#2._Mostrar primeras filas del dataset
robcsv.head()
```

```python
# Ver las filas que tienen al menos un valor nulo
robcsv_nulls = robcsv[robcsv.isnull().any(axis=1)]
robcsv_nulls
```

```python
#3._Limpiar los datos
#Eliminar filas con valores nulos
robcsv=robcsv.dropna()

#Filtrar valores atípicos (por ejemplo, si la velocidad es negativa)
robcsv.head()
```

```python
#4._ Normalizar los datos numéricos con Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Seleccionar las columnas numéricas del DataFrame
columnasnor = robcsv.select_dtypes(include=['float64', 'int64']).columns

# Aplicar el escalado Min-Max
robcsv[columnasnor] = scaler.fit_transform(robcsv[columnasnor])

# Ver las primeras filas del DataFrame normalizado
robcsv.head()
```
